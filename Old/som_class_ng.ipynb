```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
som = pd.read_csv('som.csv')
```

```python
# rename columns: replace R with som, replace '.' with '_'
som.columns = som.columns.str.replace('R', 'som').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
som = som.rename(columns={ som.columns[0]: 'sample_no'})
som
```

```python
# Replace all values with group_number
values = 'Group_' + som.iloc[:, 1:].astype(str)
values
```

```python
lithic_cat = lithic['Category']
lithic_cat
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values
```

```python
# add sample_no from som to values
values['sample_no'] = som['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as som
som = values
som
```

```python
# som1_2, som2_2, som3_2, som4_2, som5_2 are all same
som1_2_cm = pd.crosstab(som.category, som.som1_2)
# find the percentage of agreement between original categories vs largest count
som1_2_cm['percentage'] = som1_2_cm.max(axis=1) / lithic_cat.value_counts()
som1_2_cm
```
