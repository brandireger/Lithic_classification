```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
km = pd.read_csv('km.csv')
```

```python
# rename columns: replace R with km, replace '.' with '_'
km.columns = km.columns.str.replace('R', 'km').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
km = km.rename(columns={ km.columns[0]: 'sample_no'})
km.head()
```

```python
# Replace all values with group_number
values = 'Group_' + km.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Category']
lithic_cat.head()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values.head()
```

```python
# add sample_no from km to values
values['sample_no'] = km['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as km
km = values
km.head()
```

```python
# km1_2, km2_2, km3_2, km4_2, km5_2 are all same
km1_2_cm = pd.crosstab(km.category, km.km1_2)
# find the percentage of agreement between original categories vs largest count
km1_2_cm['percentage'] = km1_2_cm.max(axis=1) / lithic_cat.value_counts()
km1_2_cm
```
