```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
cmeans = pd.read_csv('cmeans.csv')
```

```python
# rename columns: replace R with cmeans, replace '.' with '_'
cmeans.columns = cmeans.columns.str.replace('R', 'cmeans').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
cmeans = cmeans.rename(columns={ cmeans.columns[0]: 'sample_no'})
cmeans.head()
```

```python
# Replace all values with group_number
values = 'Group_' + cmeans.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Category']
lithic_cat.head()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values.head()
```

```python
# add sample_no from cmeans to values
values['sample_no'] = cmeans['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as cmeans
cmeans = values
cmeans.head()
```

```python
# cmeans1_2, cmeans2_2, cmeans3_2, cmeans4_2, cmeans5_2 are all same
cmeans1_2_cm = pd.crosstab(cmeans.category, cmeans.cmeans1_2)
# find the percentage of agreement between original categories vs largest count
cmeans1_2_cm['percentage'] = cmeans1_2_cm.max(axis=1) / lithic_cat.value_counts()
cmeans1_2_cm
```
