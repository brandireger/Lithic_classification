```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
gmm = pd.read_csv('gmm.csv')
```

```python
# rename columns: replace R with gmm, replace '.' with '_'
gmm.columns = gmm.columns.str.replace('R', 'gmm').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
gmm = gmm.rename(columns={ gmm.columns[0]: 'sample_no'})
gmm.head()
```

```python
# Replace all values with group_number
values = 'Group_' + gmm.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Category']
lithic_cat.head()
```

```python
lithic_cat.value_counts()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values.head()
```

```python
# add sample_no from gmm to values
values['sample_no'] = gmm['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as gmm
gmm = values
gmm.head()
```

```python
# gmm1_2, gmm2_2, gmm3_2, gmm4_2, gmm5_2 are all same
gmm1_2_cm = pd.crosstab(gmm.category, gmm.gmm1_2)
# find the percentage of agreement between original categories vs largest count
gmm1_2_cm['percentage'] = gmm1_2_cm.max(axis=1) / lithic_cat.value_counts()
gmm1_2_cm
```
