```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
lee = pd.read_csv('lee.csv')
```

```python
# rename columns: replace R with lee, replace '.' with '_'
lee.columns = lee.columns.str.replace('R', 'lee').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
lee = lee.rename(columns={ lee.columns[0]: 'sample_no'})
lee.head()
```

```python
# Replace all values with group_number
values = 'Group_' + lee.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Category']
lithic_cat.head()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values.head()
```

```python
# add sample_no from lee to values
values['sample_no'] = lee['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as lee
lee = values
lee.head()
```

```python
# lee1_2, lee2_2, lee3_2, lee4_2, lee5_2 are all same
lee1_2_cm = pd.crosstab(lee.category, lee.lee1_2)
# find the percentage of agreement between original categories vs largest count
lee1_2_cm['percentage'] = lee1_2_cm.max(axis=1) / lithic_cat.value_counts()
lee1_2_cm
```
