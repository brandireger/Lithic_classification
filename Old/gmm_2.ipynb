```python
import pandas as pd

lithic = pd.read_csv('AllSamples.csv')
gmm = pd.read_csv('gmm_2.csv')
```

```python
# rename 'Unnamed: 0' to 'sample_no'
gmm = gmm.rename(columns={ gmm.columns[0]: 'sample_no'})
gmm.head()
```

```python
# Replace all values with group_number
values = 'Group_' + gmm.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Vis']
lithic_cat.head()
```

```python
known = lithic['is_known']
known.head()
```

```python
lithic_cat.value_counts()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values['known'] = known
values.head()
```

```python
# add sample_no from gmm to values
values['sample_no'] = gmm['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-3:] + cols[:-3]
values = values[cols]
values.head()
```

```python
# set index to first column
values = values.set_index('sample_no')
# save values as gmm
gmm = values
gmm.head()
```

```python
# break gmm into kown vs guesses
gmm_known = gmm[gmm.known != 'Guess']
gmm_guess = gmm[gmm.known == 'Guess']

gmm_known.head()
```

```python
gmm2_cm = pd.crosstab(gmm_known.category, gmm_known['2'])
# find the percentage of agreement between original categories vs largest count
gmm2_cm['percentage'] = gmm2_cm.max(axis=1) / gmm_known.category.value_counts()
gmm2_cm
```

```python
gmm2_cm_g = pd.crosstab(gmm_guess.category, gmm_guess['2'])
# find the percentage of agreement between original categories vs largest count
gmm2_cm_g['percentage'] = gmm2_cm_g.max(axis=1) / gmm_guess.category.value_counts()
gmm2_cm_g
```
