```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
sc = pd.read_csv('sc.csv')
```

```python
# rename columns: replace R with sc, replace '.' with '_'
sc.columns = sc.columns.str.replace('R', 'sc').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
sc = sc.rename(columns={ sc.columns[0]: 'sample_no'})
sc.head()
```

```python
# Replace all values with group_number
values = 'Group_' + sc.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Category']
lithic_cat.head()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values.head()
```

```python
# add sample_no from sc to values
values['sample_no'] = sc['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as sc
sc = values
sc.head()
```

```python
# sc1_2, sc2_2, sc3_2, sc4_2, sc5_2 are all same
sc1_2_cm = pd.crosstab(sc.category, sc.sc1_2)
# find the percentage of agreement between original categories vs largest count
sc1_2_cm['percentage'] = sc1_2_cm.max(axis=1) / lithic_cat.value_counts()
sc1_2_cm
```
