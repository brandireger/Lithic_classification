```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
ap = pd.read_csv('ap.csv')
```

```python
# rename columns: replace R with gmm, replace '.' with '_'
ap.columns = ap.columns.str.replace('R', 'ap').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
ap = ap.rename(columns={ ap.columns[0]: 'sample_no'})
ap.head()
```

```python
# Replace all values with group_number
values = 'Group_' + ap.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Category']
lithic_cat.head()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values.head()
```

```python
# add sample_no from gmm to values
values['sample_no'] = ap['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as gmm
ap = values
ap.head()
```

```python
# ap1_2, ap2_2, ap3_2, ap4_2, ap5_2 are all same
ap1_2_cm = pd.crosstab(ap.category, ap.ap1_2)
# find the percentage of agreement between original categories vs largest count
ap1_2_cm['percentage'] = ap1_2_cm.max(axis=1) / lithic_cat.value_counts()
ap1_2_cm
```
