```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
pam = pd.read_csv('pam.csv')
```

```python
# rename columns: replace R with pam, replace '.' with '_'
pam.columns = pam.columns.str.replace('R', 'pam').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
pam = pam.rename(columns={ pam.columns[0]: 'sample_no'})
pam.head()
```

```python
# Replace all values with group_number
values = 'Group_' + pam.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Category']
lithic_cat.head()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values.head()
```

```python
# add sample_no from pam to values
values['sample_no'] = pam['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as pam
pam = values
pam.head()
```

```python
# pam1_2, pam2_2, pam3_2, pam4_2, pam5_2 are all same
pam1_2_cm = pd.crosstab(pam.category, pam.pam1_2)
# find the percentage of agreement between original categories vs largest count
pam1_2_cm['percentage'] = pam1_2_cm.max(axis=1) / lithic_cat.value_counts()
pam1_2_cm
```
