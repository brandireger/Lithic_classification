```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
block = pd.read_csv('block.csv')
```

```python
# rename columns: replace R with block, replace '.' with '_'
block.columns = block.columns.str.replace('R', 'block').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
block = block.rename(columns={ block.columns[0]: 'sample_no'})
block.head()
```

```python
# Replace all values with group_number
values = 'Group_' + block.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Category']
lithic_cat.head()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values.head()
```

```python
# add sample_no from block to values
values['sample_no'] = block['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as block
block = values
block.head()
```

```python
# block1_2, block2_2, block3_2, block4_2, block5_2 are all same
block1_2_cm = pd.crosstab(block.category, block.block1_2)
# find the percentage of agreement between original categories vs largest count
block1_2_cm['percentage'] = block1_2_cm.max(axis=1) / lithic_cat.value_counts()
block1_2_cm
```
