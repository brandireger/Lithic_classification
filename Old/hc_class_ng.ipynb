```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
hc = pd.read_csv('hc.csv')
```

```python
# rename columns: replace R with hc, replace '.' with '_'
hc.columns = hc.columns.str.replace('R', 'hc').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
hc = hc.rename(columns={ hc.columns[0]: 'sample_no'})
hc.head()
```

```python
# Replace all values with group_number
values = 'Group_' + hc.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Category']
lithic_cat.head()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values.head()
```

```python
# add sample_no from hc to values
values['sample_no'] = hc['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as hc
hc = values
hc.head()
```

```python
# hc1_2, hc2_2, hc3_2, hc4_2, hc5_2 are all same
hc1_2_cm = pd.crosstab(hc.category, hc.hc1_2)
# find the percentage of agreement between original categories vs largest count
hc1_2_cm['percentage'] = hc1_2_cm.max(axis=1) / lithic_cat.value_counts()
hc1_2_cm
```
