```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
diana = pd.read_csv('diana.csv')
```

```python
# rename columns: replace R with diana, replace '.' with '_'
diana.columns = diana.columns.str.replace('R', 'diana').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
diana = diana.rename(columns={ diana.columns[0]: 'sample_no'})
diana
```

```python
# Replace all values with group_number
values = 'Group_' + diana.iloc[:, 1:].astype(str)
values
```

```python
lithic_cat = lithic['Category']
lithic_cat
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values
```

```python
# add sample_no from diana to values
values['sample_no'] = diana['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as diana
diana = values
diana
```

```python
# diana1_2, diana2_2, diana3_2, diana4_2, diana5_2 are all same
diana1_2_cm = pd.crosstab(diana.category, diana.diana1_2)
# find the percentage of agreement between original categories vs largest count
diana1_2_cm['percentage'] = diana1_2_cm.max(axis=1) / lithic_cat.value_counts()
diana1_2_cm
```
