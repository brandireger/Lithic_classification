```python
import pandas as pd

lithic = pd.read_csv('Lithic_Database.csv')
brunet = pd.read_csv('brunet.csv')
```

```python
# rename columns: replace R with brunet, replace '.' with '_'
brunet.columns = brunet.columns.str.replace('R', 'brunet').str.replace('.', '_')
# rename 'Unnamed: 0' to 'sample_no'
brunet = brunet.rename(columns={ brunet.columns[0]: 'sample_no'})
brunet.head()
```

```python
# Replace all values with group_number
values = 'Group_' + brunet.iloc[:, 1:].astype(str)
values.head()
```

```python
lithic_cat = lithic['Category']
lithic_cat.head()
```

```python
# problem: NaN at end because of mismatch in index
values['category'] = lithic_cat
values.head()
```

```python
# add sample_no from brunet to values
values['sample_no'] = brunet['sample_no']
# move category column to front
cols = values.columns.tolist()
cols = cols[-2:] + cols[:-2]
values = values[cols]
# set index to first column
values = values.set_index('sample_no')
# save values as brunet
brunet = values
brunet.head()
```

```python
# brunet1_2, brunet2_2, brunet3_2, brunet4_2, brunet5_2 are all same
brunet1_2_cm = pd.crosstab(brunet.category, brunet.brunet1_2)
# find the percentage of agreement between original categories vs largest count
brunet1_2_cm['percentage'] = brunet1_2_cm.max(axis=1) / lithic_cat.value_counts()
brunet1_2_cm
```
